##**Cleaning data**

No Negative records if the user doesnt leave any negative feedback comment, and No Positive by default for positive comments. We need to remove these parts from our texts.

```python
reviews_df["review"] = reviews_df["review"].apply(lambda x: x.replace("No Negative", "").replace("No Positive", ""))
```

In the follwing we want to clean the text data with various operations as below:

```python
import string
from nltk import pos_tag
from nltk.corpus import stopwords
from nltk.tokenize import WhitespaceTokenizer
from nltk.stem import WordNetLemmatizer
import nltk
from nltk.corpus import wordnet
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
```

```python
# return the wordnet object value corresponding to the POS tag
from nltk.corpus import wordnet

def get_wordnet_pos(pos_tag):
    if pos_tag.startswith('J'):
        return wordnet.ADJ
    elif pos_tag.startswith('V'):
        return wordnet.VERB
    elif pos_tag.startswith('N'):
        return wordnet.NOUN
    elif pos_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

The function below we are using for preparing text data for natural language processing.

```python
def clean_text(text):
    #Converts the text to lowercase.
    text = text.lower()
    #Tokenizes the text and removes punctuation.
    text = [word.strip(string.punctuation) for word in text.split(" ")]
    #Removes words that contain numbers
    text = [word for word in text if not any(c.isdigit() for c in word)]
    #Removes stop words (common words such as the, a, an, etc.).
    stop = stopwords.words('english')
    text = [x for x in text if x not in stop]
    #Removes empty tokens.
    text = [t for t in text if len(t) > 0]
    #Part-of-speech tags the text.
    pos_tags = pos_tag(text)
    text = [WordNetLemmatizer().lemmatize(t[0], get_wordnet_pos(t[1])) for t in pos_tags]
    #Removes words with only one letter.
    text = [t for t in text if len(t) > 1]
    #Joins all the remaining tokens into a single string.
    text = " ".join(text)
    return(text)
```

Part Of Speech (POS) tagging: assign a tag to every word to define if it corresponds to a noun, a verb etc. using the WordNet lexical database
lemmatize the text: transform every word into their root form (e.g. beds -> bed, booked -> book)

```python
reviews_df["review_clean"] = reviews_df["review"].apply(lambda x: clean_text(x))
```

After cleaning our data, we can proceed with feature engineering for our modelization part.

##**Feature engineering**

In this section, we begin by incorporating sentiment analysis features. We utilize Vader, a sentiment analysis tool that is part of the NLTK module. Vader uses a lexicon of words to identify positive and negative sentiments and considers the context of sentences to determine sentiment scores. For each text, Vader returns four values: a **neutrality score**, a **positivity score**, a **negativity score**, and an **overall score** that summarizes the previous scores.

```python
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
reviews_df["sentiments"] = reviews_df["review"].apply(lambda x: sid.polarity_scores(x))
reviews_df = pd.concat([reviews_df.drop(['sentiments'], axis=1), reviews_df['sentiments'].apply(pd.Series)], axis=1)
```

We decided to add some metrics for each text:
1. number of characters in the text(**nb_chars**)
2. number of words in the text(**nb_words**)

```python
reviews_df["nb_chars"] = reviews_df["review"].apply(lambda x: len(x))
reviews_df["nb_words"] = reviews_df["review"].apply(lambda x: len(x.split(" ")))
```

The next step is to extract vector representations for each review. For that we use **Gensim module** that creates a numerical vector representation of every word in the corpus by using the contexts in which they appear (Word2Vec) with shallow neural networks. **The interesting part is that similar words will have similar representation vectors.**



```python
from gensim.test.utils import common_texts
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
```

We can also transform each text into numerical vectors using the word vectors (Doc2Vec). To do this, we first need to train a Doc2Vec model by feeding in our text data.

```python
documents = [TaggedDocument(doc, [i]) for i, doc in enumerate(reviews_df["review_clean"].apply(lambda x: x.split(" ")))]
model = Doc2Vec(documents, vector_size=5, window=2, min_count=1, workers=4)
doc2vec_df = reviews_df["review_clean"].apply(lambda x: model.infer_vector(x.split(" "))).apply(pd.Series)
doc2vec_df.columns = ["doc2vec_vector_" + str(x) for x in doc2vec_df.columns]
reviews_df = pd.concat([reviews_df, doc2vec_df], axis=1)
```

To improve the accuracy of our analysis, we incorporate the TF-IDF (Term Frequency - Inverse Document Frequency) values for every word and document. This metric is more effective than simply counting the number of times each word appears in every document because it takes into account the relative importance of words in the texts. Words that appear in almost every text are unlikely to provide useful information for analysis, while rare words may have more meanings.


```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
tfidf = TfidfVectorizer(min_df = 20)
tfidf_result = tfidf.fit_transform(reviews_df["review_clean"]).toarray()
tfidf_df = pd.DataFrame(tfidf_result, columns=tfidf.get_feature_names_out())
tfidf_df.columns = ["word_" + str(x) for x in tfidf_df.columns]
tfidf_df.index = reviews_df.index
reviews_df = pd.concat([reviews_df, tfidf_df], axis=1)
reviews_df.head()
```

The TF-IDF metric solves this problem by computing the classic number of times the word appears in the text (TF) and the relative importance of this word, which depends on how many texts the word can be found (IDF). We add TF-IDF columns for every word that appears in at least 20 different texts to filter some of them and reduce the size of the final output.

##**Model**


Here we use a Random Forest (RF) classifier for our predictions.



```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
label = "is_bad_review"
ignore_cols = [label, "review", "review_clean"]
features = [c for c in reviews_df.columns if c not in ignore_cols]
X_train, X_test, y_train, y_test = train_test_split(reviews_df[features], reviews_df[label], test_size = 0.20, random_state = 42)
```

```python
rf = RandomForestClassifier(n_estimators = 100, random_state = 42)
rf.fit(X_train, y_train)
feature_importances_df = pd.DataFrame({"feature": features, "importance": rf.feature_importances_}).sort_values("importance", ascending = True)
feature_importances_df.head(15)
```

The sentiment analysis features computed earlier are indeed the most important ones. The vector representations of the texts are also crucial in our training. Additionally, some words have a significant impact on the models performance.

###**ROC curve**

The ROC curve is usually a good graph to summarize the quality of our classifier. The higher the curve is above the diagonal baseline, the better the predictions.

```python
from sklearn.metrics import roc_curve, auc, roc_auc_score
import matplotlib.pyplot as plt

y_pred = [x[1] for x in rf.predict_proba(X_test)]
fpr, tpr, thresholds = roc_curve(y_test, y_pred, pos_label = 1)
roc_auc = auc(fpr, tpr)
plt.figure(1, figsize = (15, 10))
lw = 2
plt.plot(fpr, tpr, color='red',
         lw=lw, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc="lower right")
plt.show()
```

However, we should **not** use the ROC curve to assess the quality of our model in this case, despite the AUC ROC (Area Under the Curve ROC) being very good. In this case, the # Negatives corresponds to the number of good reviews, which is very high because our dataset is imbalanced. This means that even with some False Positives, our FPR will tend to stay very low. Our model will be able to make many false positive predictions and still have a low false positive rate, while increasing the true positive rate and therefore artificially increasing the AUC ROC metric.

A better metric in this imbalanced situation is the AUC PR (Area Under the Curve Precision Recall), or also called AP (Average Precision).

###**PR curve**

```python
from sklearn.metrics import average_precision_score, precision_recall_curve
import matplotlib.pyplot as plt
import inspect
```

```python
average_precision = average_precision_score(y_test, y_pred)
precision, recall, _ = precision_recall_curve(y_test, y_pred)
step_kwargs = ({'step': 'post'}
               if 'step' in inspect.signature(plt.fill_between).parameters
               else {})
```

```python
plt.figure(1, figsize=(15, 10))
plt.step(recall, precision, color='b', alpha=0.2, where='post')
plt.fill_between(recall, precision, alpha=0.2, color='r', **step_kwargs)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.0, 1.05])
plt.xlim([0.0, 1.0])
plt.title('2-class Precision-Recall curve: AP={0:0.2f}'.format(average_precision))
plt.show()
```

The precision decreases as the recall increases. This indicates that we need to choose a prediction threshold that is tailored to our needs. If our objective is to achieve a high recall, we should set a low prediction threshold that will enable us to detect most of the positive class observations, but with a low precision. Conversely, if we want to be highly confident in our predictions and are willing to miss some positive observations, we should set a high threshold that will yield high precision and low recall.

At the end it is completely possible to use only raw text as input for making predictions. The most important thing is to be able to extract the relevant features from this raw source of data.

#**Recommender System** (by  country)


In this phase we're trying to build a recommender system, which will give the users hotels available in their destination sorted in the order that they have chosen. The base for sorting hotels in order is their average ratings. Our dataset doesn't include this metric, and we have to calculate it ourselves.

First, we get a copy from our dataframe, so in case o any unexpected failure we can always woll back and won't lose our data.

```python
recomm_df = hotel_df.copy(deep=True)
```

```python
recomm_df.head(7)
```

```python
hotel_address= recomm_df['Hotel_Address'].values.tolist()
print(set(hotel_address))
```

As you can see the last part of the hotel's address is the name of the country. Since our recommender system should recommend hotels of the user's destination, we have to extract the country from address and store it as an additional column in our dataset.

```python
countries=[]
for address in hotel_address:
  splitted_address = address.split()
  country = splitted_address[-1]
  countries.append(country)

recomm_df['country'] = countries
recomm_df.head(8)
```

Now , we have to actually calculate the average rating of each hotel and store it in a new column named 'average_rating'.

```python
recomm_df['average_rating'] = recomm_df.groupby('Hotel_Name')['Reviewer_Score'].transform('mean')
recomm_df.head()
```

```python
recomm_df[recomm_df['country']=='Austria']
```

We can leave the choice for the order of hotels being listed to users. So whenever a user wants to get hotel recommendations, they should first enter their desttination country and then the order they'd prefer.

```python
def get_input():
  Destination = input("Please enter the country of your destination: ")
  Order = int(input("In what order do you like the list of hotels to be shown : 1) Ascending 2) Descending"))
  return Destination, Order
```

```python
def get_result(Destination , Order):
  country_hotel = recomm_df[recomm_df['country'] == Destination]
  if(Order == 1):
    country_hotel = country_hotel.sort_values(by='average_rating',ascending=True)
    hotel_names = country_hotel['Hotel_Name'].values.tolist()
  else:
    country_hotel = country_hotel.sort_values(by='average_rating',ascending=True)
    hotel_names = country_hotel['Hotel_Name'].values.tolist()
    hotel_names=hotel_names[::-1]

  print(list(dict.fromkeys(hotel_names)))
```

```python
def continuation():
  choice = int(input("Do you want to continue?  1)Yes 2)No"))
  return choice
```

Here's one instance of the recommender system working:

```python
while(True):
  Destination,Order = get_input()
  get_result(Destination,Order)
  choice = continuation()
  if( choice == 2):
      break;
```
